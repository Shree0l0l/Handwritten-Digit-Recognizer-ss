```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
(X_train,y_train),(X_test,y_test)=keras.datasets.mnist.load_data()
```

```python
#Scaling improves the accuracy of the machine learning model
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

```python
X_train_flattened=X_train.reshape(len(X_train),28*28)
X_test_flattened=X_test.reshape(len(X_test),28*28)
X_test_flattened.shape
```

```python
X_test_flattened[0]
```

```python
#Creating a neural network
#Stack of layers in our neural network thus accepts every layer as an element
model=keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid')    #Dense-every neuron in the first layer is connected to every other neuron in the second layer
    
])
#Optimizer helps to train efficiently when back propogation or training goes on and allow to reach global optima in an efficient way
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  #Computes the cross entropy loss btw the labels and the predictions ,mean squared error can also work as well
    metrics=['accuracy']
)
#training
model.fit(X_train_flattened,y_train,epochs=5) #Epoch is the number of iteration required for the training
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
plt.figure(figsize=(4,4))
plt.imshow(X_test[0])
plt.axis('off')
plt.show()
```

```python
y_predicted=model.predict(X_test_flattened)
y_predicted[0]
```

```python
print(np.argmax(y_predicted[0]))  #prints the index of the maximum value in the array(which have highest accuracy match 
```

```python
plt.figure(figsize=(4,4))
plt.imshow(X_test[1])
plt.axis('off')
plt.show()
```

```python
y_predicted=model.predict(X_test_flattened)
print(np.argmax(y_predicted[1]))
```

```python
plt.figure(figsize=(4,4))
plt.imshow(X_test[3])
plt.axis('off')
plt.show()
```

```python
print(np.argmax(y_predicted[3]))
```

```python
y_predicted_labels=[np.argmax(i) for i in y_predicted]  #Because the y_predicted is in floating point so every values are changed to corresponding number
y_predicted_labels[:5]
```

```python
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Accurate')
plt.show()
#It tells that 967 times our label was 0 and our model predicted it correctly to be 0
#1107 it was 1 and our model predicted it to be 1
#9 times our model was 2 but our model predicted it to be 
#Not in the diagonal means errors
```

```python
#Adding a hidden Layer to improve the performance
# we can add more hidden layers
model=keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
     keras.layers.Dense(10,activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy', 
    metrics=['accuracy']
)

model.fit(X_train_flattened,y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

```python
y_predicted=model.predict(X_test_flattened)
y_predicted_labels=[np.argmax(i) for i in y_predicted] 
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Accurate')
plt.show()
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,activation='relu'),
     keras.layers.Dense(10,activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy', 
    metrics=['accuracy']
)

model.fit(X_train,y_train,epochs=5)
```

```python
model.evaluate(X_test,y_test)
```
